```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
import random as rd

doorsCons = np.array([0,0,0])
doorsSwitch = np.array([0,0,0])
doorsNew = np.array([0,0,0])
winSwitch = 0
winNew = 0
winCons = 0
giocate = 100
def clean():
    for i in range (3):
        doorsCons[i] = 0
        doorsSwitch[i] = 0
        doorsNew[i] = 0
def checkGoat(array):
    i=0
    while array[i] == 0 : 
        i+=1
    return i;
giocate = 100
#PRIMA DECISIONE

for i in range(giocate):
    
    #decido dove si trova la macchina e quale porta decide il conservatore e lo switcher
    carCons = rd.randint(0,2)
    dec = rd.randint(0,2)
    carSwitch = rd.randint(0,2)
    carNew = rd.randint(0,2)
    doorsCons[carCons] = 1
    doorsCons[dec] += 2
    doorsSwitch[carSwitch] = 1
    doorsSwitch[dec] +=2
    doorsNew[carNew] = 1
    
    #decido se il conservatore ha vinto o no
    
    if(doorsCons[dec]==3):
        print("conservatore vince")
        winCons += 1
    else:
        print("conservatore perde")
        
#SECONDA DECISIONE      
    #controllo in che porta (non scelta) si trova la capra dello Switch
    GoatSwitch = checkGoat(doorsSwitch)
    
    #cambio decisione switch tra le due rimanenti
    doorsSwitch[dec] += -2
    if ((GoatSwitch + 1)%3 == dec):
        doorsSwitch[(GoatSwitch + 2)%3] += 2 
        choice = (GoatSwitch + 2)%3
    else:
        doorsSwitch[(GoatSwitch + 1)%3] += 2
        choice = (GoatSwitch + 1)%3
        
    #printo risultato switch
#    if (doorsSwitch[choice] == 3):
#    print("switch vince")
#        winSwitch += 1
#    else:
#        print("switch perde")
    for i in range (3):
        print(doorsSwitch[i])
    print("------S-----")
    #controllo dove si trova la capra del New    
    GoatNew = checkGoat(doorsNew)
    Choice_New = rd.randint(1,2)
    doorsNew[(GoatNew + Choice_New) %3] += 2
#    if (doorsNew[(GoatNew + Choice_New) %3] == 3):
#        print("new vince")
#        winNew += 1
#    else:
#        print("new perde")
#    print("----------")
    for i in range (3):
        print(doorsSwitch[i])
    print("------N-----")
    clean()
    
print(winCons/giocate)
print(winSwitch/giocate)
print(winNew/giocate)
```
